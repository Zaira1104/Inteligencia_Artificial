
##Analizando la relacion entre las caracteristicas con la probabilidad de supervivencia/mortalidad del TITANIC

```python
df.columns
```

```python
def bar_chart(feature):
  survived=df[df['Survived']==1][feature].value_counts()
  dead=df[df['Survived']==0][feature].value_counts()
  test=pd.DataFrame([survived,dead])
  test.index=['Survived','Dead']
  test.plot(kind='bar',stacked=True,figsize=(10,5))
```

```python
bar_chart("Sex")
```

La grafica seala que las mujer tiene casi 3 veces ms posibilides de sobrevivir que los hombre.


```python
bar_chart("Pclass")
```


La grafica seala que la clase alta tuvo mayot probabilidad de supervivencia con respescto a
