```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt #para graficas
import seaborn as sns #mejora las graficas
```

```python
data = pd.read_excel("/content/Salary.xlsx")
data.head()
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
x=df["YearsExperience"]
y=df["Salary"]

plt.figure(figsize=(10,6))
plt.scatter(x,y, color ="blue", label="Datos Reales")
plt.title("Salario vs Años de Experiencia")
plt.xlabel("Años de Experiencia")
plt.ylabel("Salario")
plt.legend()
plt.grid(True)
plt.show()
```

```python
model=LinearRegression()
model.fit(x.values.reshape(-1, 1),y)

y_pred=model.predict(x.values.reshape(-1, 1))

r2=r2_score(y,y_pred)
print("Coeficiente de Determinación (R^2):", r2)
```

```python
plt. figure(figsize=[10, 6])
plt.scatter (x, y, color="blue",label="Datos Reales")
plt.plot(x,y_pred, color="red", linewidth=3, label= "Regresión Lineal")
plt.xlabel("Años de experiencia")
plt. ylabel("Salario")
plt. title("Salario vs Años de experiencia")
plt. legend()
plt.grid (True)
plt. show()

print(f"Ecuacion de la recta: y={model.coef_[0]:2f}x+{model.intercept_:2f}")
print(f"Coeficiente de Determinación (R^2): {r2:2f}")
```
